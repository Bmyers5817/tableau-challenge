```python
# Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
import seaborn as sn
```

```python
sep2020 = pd.read_csv("202009-citibike-tripdata.csv")
oct2020 = pd.read_csv("202010-citibike-tripdata.csv")
nov2020 = pd.read_csv("202011-citibike-tripdata.csv")
dec2020 = pd.read_csv("202012-citibike-tripdata.csv")
jan2021 = pd.read_csv("202101-citibike-tripdata.csv")
feb2021 = pd.read_csv("202102-citibike-tripdata.csv")
mar2021 = pd.read_csv("202103-citibike-tripdata.csv")
apr2021 = pd.read_csv("202104-citibike-tripdata.csv")
```

```python
citibike_df = pd.concat([sep2020, oct2020, nov2020, dec2020, jan2021, feb2021, mar2021, apr2021])
citibike_df.head(5)
```

```python
citibike_df['gender']=citibike_df['gender'].replace([0, 1, 2],['Unknown', 'Male', 'Female'])
```
